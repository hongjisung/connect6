```python
import pandas as pd
from pandas import Series, DataFrame
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("data2.csv")
```

```python
df
```

```python
col = df.columns
col = [i.strip() for i in col]
df.columns = col
```

```python
x_data = df[col[:-1]]
y_data = df.pop("result")
```

```python
y_data.replace({" WIN":1, " LOSE":0, "WIN":1, "LOSE":0}, inplace=True)
```

```python
lr = LogisticRegression()
lr.fit(x_data,y_data)
```

```python
# make the one hot encoded data to find the parameter
test = [[0  for j in range(0,len(col)-1)] for i in range(0,len(col)-1)]
for i in range(0,len(col)-1):
    test[i][i] = 1
dt = pd.DataFrame(test)
```

```python
params = list(lr.decision_function(dt))
```

```python
f = open('save.txt', 'w')
f.write('{')
for p in params:
    f.write(str(p))
    f.write(',')
f.write('}')
f.close()
```
